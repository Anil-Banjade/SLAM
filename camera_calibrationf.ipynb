```python
#Else start from here if you have frames
# MOST IMPORTANT STEP IS TO SET square_size in code cell below this cell
#also set the path here
import os
path = "/content/drive/My Drive/calibration_images/good_frames"
files = os.listdir(path)
print(files)
```

```python
square_size = 26  # in millimeters( the size of square, please measure with scale if necessary this needs to be accurate)
CHECKERBOARD = (8,6)  # 8 inner row corner, 6 inner coloumn corner(count inner corners only and set it to here, not outer total corners)
```

```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt




#termination criteria for corner refinement: subpixel accuracy for better localization of the exact corner
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

#prepare 3D object points
objp = np.zeros((CHECKERBOARD[0]*CHECKERBOARD[1], 3), np.float32)
objp[:, :2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
objp *= square_size  # scale to real-world units

objpoints = []  #3D points
imgpoints = []  #2D points

# image_paths = glob.glob(path + "/*.jpg")
# path = "/content/drive/My Drive/calibration_images/new"
for path_ in files:
    img_path = os.path.join(path, path_)

    img = cv2.imread(img_path)
    print("boya")
    print(path_)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD, None)

    if ret:
        objpoints.append(objp)
        corners2 = cv2.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)
        imgpoints.append(corners2)


        img = cv2.drawChessboardCorners(img, CHECKERBOARD, corners2, ret)
        plt.figure(figsize=(8,6))
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(os.path.basename(path))
        plt.axis('off')
        plt.show()
```

```python
print("Number of valid images with detected corners:", len(objpoints))

print("Shape of last gray image:", gray.shape[::-1])
ret, K, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)

print("Calibration successful")
print("\nCamera Matrix (K):\n", K)
print("\nDistortion Coefficients:\n", dist.ravel())
print("rvec: ", rvecs)
# HERER LIES CAMERA Intrisics:
```
